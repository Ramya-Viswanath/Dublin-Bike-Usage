```python
import datetime as dt
import numpy as np
import datetime as dt
def bin_time(x):
    if x.time() < dt.time(6):
        return "Overnight"
    elif x.time() < dt.time(11):
        return "6AM-10AM"
    elif x.time() < dt.time(16):
        return "11AM-3PM"
    elif x.time() < dt.time(20):
        return "4PM-7PM"
    elif x.time() <= dt.time(23):
        return "8PM-11PM"
    else:
        return "Overnight"

pre_pandemic_data.sort_values(by=['STATION ID','LAST UPDATED'], inplace=True)
pre_pandemic_data['OCCUPANCY'] =  pre_pandemic_data['AVAILABLE BIKES'] / pre_pandemic_data['BIKE STANDS']
pre_pandemic_data['FULL'] = np.where(pre_pandemic_data['OCCUPANCY'] == 0, 1, 0)
pre_pandemic_data['EMPTY'] = np.where(pre_pandemic_data['OCCUPANCY'] == 1, 1, 0)
pre_pandemic_data['DAY_NUMBER'] = pre_pandemic_data['TIME'].dt.dayofweek
pre_pandemic_data['DAY_TYPE'] = np.where(pre_pandemic_data['DAY_NUMBER'] <= 4, 'Weekday', (np.where(pre_pandemic_data['DAY_NUMBER'] == 5, 'Saturday', 'Sunday')))
pre_pandemic_data["TIME_TYPE"] = pre_pandemic_data['TIME'].apply(bin_time)
pre_pandemic_data['HOUR'] = pre_pandemic_data['TIME'].dt.hour
pre_pandemic_data['MONTH'] = pre_pandemic_data['TIME'].dt.month
print(pre_pandemic_data.sample(5))
```
